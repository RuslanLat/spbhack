```python
import difflib
```

```python
def similarity(s1, s2='аптерский 18 спб'):
  normalized1 = s1.lower()
  normalized2 = s2.lower()
  matcher = difflib.SequenceMatcher(None, normalized1, normalized2)
  return matcher.real_quick_ratio()
```

```python
building['similarity'] = building['full_address'].map(similarity)
```

```python
building.sort_values(by='similarity', ascending=False)['full_address'][:10]
```

```python
from pyjarowinkler import distance
```

```python
def similarity(s1, s2='аптерский 18 спб'):
  normalized1 = s1.lower()
  normalized2 = s2.lower()
  ratio = distance.get_jaro_distance(normalized1, normalized2, winkler=True, scaling=0.1)
  return ratio
```

```python
building['pyjarowinkler'] = building['full_address'].map(similarity)
```

```python
building.sort_values(by='pyjarowinkler', ascending=False)['full_address'][:10]
```
